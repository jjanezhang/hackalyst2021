## Scoring

#### clean up data

```python
df_clean = df1
del df_clean["id"]
del df_clean["duration_ms"]
del df_clean["liveness"]
del df_clean["popularity"]
del df_clean["release_date"]
del df_clean["speechiness"]
del df_clean["year"]
del df_clean["loudness"]
del df_clean["instrumentalness"]
df_clean
```

```python
def sad_score(df):
    scoreList=[]
    for ind in range(len(df)):
        sscore=0
        if (df.loc[ind,"acousticness"] > 0.9):
            sscore+=2
        if (df.loc[ind, "valence"] < 0.25):
            sscore+=3    
        if (((df.loc[ind,"key"] == 0) or (df.loc[ind,"key"] == 1) or (df.loc[ind,"key"] == 2) or
             (df.loc[ind,"key"] == 3) or (df.loc[ind,"key"] == 5)) & df.loc[ind, "mode"] == 0):
            sscore+=4
        if (df.loc[ind,"energy"] < 0.25):
            sscore+=1
        if (df.loc[ind,"tempo"] < 70):
            sscore+=2
        scoreList.append(sscore)
    return scoreList
```

```python
def happy_score(df):
    scoreList=[]
    for ind in range(len(df)):
        sscore=0
        if (df.loc[ind, "valence"] > 0.75):
            sscore+=3    
        if (((df.loc[ind,"key"] == 2) or (df.loc[ind,"key"] == 4) or 
             (df.loc[ind,"key"] == 10) or (df.loc[ind,"key"] == 7)) & df.loc[ind, "mode"] == 1):
            sscore+=4
        if (df.loc[ind,"energy"] > 0.75):
            sscore+=1
        if (df.loc[ind,"tempo"] > 150):
            sscore+=2
        scoreList.append(sscore)
    return scoreList
```

```python
def joy_score(df):
    scoreList=[]
    for ind in range(len(df)):
        sscore=0
        if (df.loc[ind, "valence"] > 0.75):
            sscore+=3    
        if (((df.loc[ind,"key"] == 0) or (df.loc[ind,"key"] == 3) or (df.loc[ind,"key"] == 7) or (df.loc[ind,"key"] == 9)
             or (df.loc[ind,"key"] == 10)) & df.loc[ind, "mode"] == 1):
            sscore+=4
        if (df.loc[ind,"energy"] > 0.55):
            sscore+=1
        if (df.loc[ind,"tempo"] > 120):
            sscore+=2
        scoreList.append(sscore)
    return scoreList
```

```python
def energetic_score(df):
    scoreList=[]
    for ind in range(len(df)):
        sscore=0
        if (df.loc[ind, "valence"] > 0.85):
            sscore+=3    
        if(((df.loc[ind,"key"] == 7)or(df.loc[ind,"key"]== 4)or(df.loc[ind,"key"] == 2)or(df.loc[ind,"key"] == 0))&df.loc[ind,"mode"]==1):
            sscore+=4
        if (df.loc[ind,"energy"] > 0.75):
            sscore+=1
        if (df.loc[ind,"tempo"] > 175):
            sscore+=2
        if (df.loc[ind, "danceability"] > 0.85):
            sscore+=1
        scoreList.append(sscore)
    return scoreList
```

```python
def angry_score(df):
    scoreList=[]
    for ind in range(len(df)):
        sscore=0
        if (df.loc[ind, "valence"] < 0.25):
            sscore+=3    
        if ((((df.loc[ind,"key"] == 3) or (df.loc[ind,"key"] == 7)) & df.loc[ind, "mode"] == 0) 
            or (((df.loc[ind,"key"] == 8) or (df.loc[ind,"key"] == 11) or (df.loc[ind,"key"] == 1)) & 
                df.loc[ind, "mode"] == 1)):
            sscore+=4
        if (df.loc[ind,"energy"] > 0.75):
            sscore+=1
        if (df.loc[ind,"tempo"] > 150):
            sscore+=2
        if (df.loc[ind,"explicit"]==1):
            sscore+=4
        scoreList.append(sscore)
    return scoreList
```

```python
sadscores=sad_score(df_clean)
hscores = happy_score(df_clean)
jscores = joy_score(df_clean)
escores = energetic_score(df_clean)
ascores = angry_score(df_clean)
```

```python
df_clean["happy"] = hscores
df_clean["joy"] = jscores
df_clean["energetic"] =escores
df_clean["sad"] = sadscores
df_clean["angry"] = ascores
```
