```python
import pandas as pd
import numpy as np
from collections import defaultdict
```

```python
df = pd.read_csv("spotify_data.csv")
```

```python
user_input = df[:100]
```

### Scoring finding second degree artists

#### Find all of users artists and rank them

```python
user_input["artist_count"] = user_input.groupby('track_artist')['track_artist'].transform('count')
user_input.sort_values(by=["artist_count"], ascending=False)
```

#### Filter out for songs with "feat"

```python
user_input["has_feat"] = df.apply(lambda x: "feat" in x["track_name"], axis=1)
```

```python
user_feat = user_input[user_input["has_feat"] == True]
```

#### Convert all songs into dict with artists by key

```python
artists_dict = defaultdict(list)

for ix, row in df.iterrows():
    track = row["track_name"]
    artists_name = row["track_artist"]
    artists_dict[artists_name].append(track)
    
```

```python
artists_dict
```

#### Return recommendations

```python
user_feat = user_feat.sort_values(by=["artist_count"], ascending=False)
```

```python
user_feat["feat_artist"] = user_feat.apply(lambda x: x["track_name"].split("feat. ")[1].split(")")[0], axis=1)
```

```python
track_recommendation = {}

for ix, row in user_feat.iterrows():
    feat_art = row["feat_artist"].split(", ")
    for artist in feat_art:
        if artists_dict[artist]:
            track_recommendation[artist] = artists_dict[artist]
```
